# Storytelling Data Visualization on Exchange Rates

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.style as style
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
```

```python
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

### About the dataset 

Dataset contains information that describes Euro dialy exchange rates between 1999 and 2021. Dataset consists of 41 columns and each of them represents the Euro's exchange rate with a specific currency (e.g. US dollar, Polish zloty, Japanese yen). Dataset contains 5699 rows and some columns have some null values. 

### Data cleaning 

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               '[Polish zloty ]': "Polish_zloty",
                               'Period\\Unit:': 'Time'},
                      inplace=True)
```

```python
exchange_rates.head()
```

```python
exchange_rates["Time"] = pd.to_datetime(exchange_rates["Time"])
```

```python
exchange_rates.sort_values("Time", inplace = True)
```

```python
exchange_rates.reset_index(drop = True, inplace = True)
```

```python
euro_to_zloty = exchange_rates[["Time", "Polish_zloty"]].copy()
```

```python
euro_to_zloty["Polish_zloty"].value_counts()
```

```python
null_values = euro_to_zloty["Polish_zloty"] == "-"
null_values.value_counts()
```

```python
euro_to_zloty = euro_to_zloty[euro_to_zloty["Polish_zloty"] != "-"]
euro_to_zloty["Polish_zloty"] = euro_to_zloty["Polish_zloty"].astype(float)
```

```python
euro_to_zloty.head()
```

```python
euro_to_zloty.tail()
```

```python
# Adding a FiveThirtyEight style
style.use("fivethirtyeight")
```

### Rolling mean

```python
plt.plot(euro_to_zloty["Time"], euro_to_zloty["Polish_zloty"], linewidth = 2)
plt.show()
```

```python
euro_to_zloty["rolling_mean_30"] = euro_to_zloty["Polish_zloty"].rolling(30).mean()
```

```python
plt.plot(euro_to_zloty["Time"], euro_to_zloty["rolling_mean_30"], linewidth = 2)
plt.show()
```

### Storytelling Data Visualisation - Poland joining European Union

```python
baseline_data = euro_to_zloty.copy()[(euro_to_zloty["Time"].dt.year >= 2001) 
                                    & (euro_to_zloty["Time"].dt.year <= 2007)]

joining_european_union = baseline_data.copy()[(baseline_data["Time"].dt.year == 2004)]

joining_european_union_specific_date = "2004-05-01"  
joining_european_union_specific_date_dt = pd.to_datetime(joining_european_union_specific_date)
```

```python
# Adding a plot 
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(baseline_data["Time"], baseline_data["rolling_mean_30"], color = "red", linewidth = 1)

# Highlighting the 2004 year
ax.plot(joining_european_union["Time"], joining_european_union["rolling_mean_30"], color = "red", linewidth = 3)

# Highlighting specific date - 01 May 2004
ax.axvline(joining_european_union_specific_date_dt, color="green", ymin = 0.05, linewidth=2, label="1 May 2004")

# Adding a legend
ax.legend()

ax.text(11000, 5.20, "EUR/PLN Exchange Rate Trends (2001–2007)",
        weight='bold')
ax.text(11000, 5.05, "The Impact of Poland's EU Accession on the Euro-PLN Exchange Rate",
        size=12)

plt.show()
```

This chart shows the EUR/PLN exchange rate trends from 2001 to 2007, highlighting Poland's accession to the EU on May 1, 2004. Before joining, the euro became more expensive, and the exchange rate peaked around 2004. After Poland joined the EU, the zloty got stronger, and the exchange rate started to slowly go down in the following years.

### Storytelling Data Visualization - Covid Pandemic Start

```python
baseline = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2016
                   ) & (euro_to_dollar['Time'].dt.year <= 2020)]

covid_pandemic = euro_to_dollar.copy()[(euro_to_dollar["Time"].dt.year >= 2020)]
```

```python
# Adding a plot
fig, ax = plt.subplots(figsize = (8,3))

ax.plot(baseline["Time"], baseline["rolling_mean_30"], 
        color = "red", linewidth = 1)

# Highlighting covid pandemic period 
ax.plot(covid_pandemic["Time"], covid_pandemic["rolling_mean_30"],
       color = "red", linewidth = 3)

# Highlighting covid pandemic start
ax.axvspan(xmin=18262, xmax=18403, ymin=0.04,
           alpha=0.3, color='grey')

# Adding a title
ax.text(16500, 1.3, "EUR/USD Exchange Rate Trends (2016–2021)",
        weight='bold')
ax.text(16500, 1.28, "Fluctuations in the Exchange Rate with the Impact of the COVID-19 Pandemic in 2020",
        size=12)


plt.show()
```

### Storytelling Data Visualization - Three United States Presidents 

```python
bush_obama_trump = euro_to_dollar.copy()[
    (euro_to_dollar["Time"].dt.year >= 2001) 
    & (euro_to_dollar["Time"].dt.year <= 2021)]
```

```python
bush = bush_obama_trump.copy()[
    (bush_obama_trump["Time"].dt.year >= 2001) 
    & (bush_obama_trump["Time"].dt.year < 2009)]
```

```python
obama = bush_obama_trump.copy()[
(bush_obama_trump["Time"].dt.year >= 2009)
& (bush_obama_trump["Time"].dt.year < 2017)]
```

```python
trump = bush_obama_trump.copy()[
(bush_obama_trump["Time"].dt.year >= 2017)
& (bush_obama_trump["Time"].dt.year < 2021)]
```

```python
plt.figure(figsize = (15,6), dpi = 100)

# Adding a subplots
ax1 = plt.subplot(2,3,1) # first subplot in a 2-row by 3-column grid, placed in the first cell (top-left corner).
ax2 = plt.subplot(2,3,2) # second subplot in a 2-row by 3-column grid, placed in the second cell (top-center) 
ax3 = plt.subplot(2,3,3) # third subplot in a 2-row by 3-column grid, placed in the third cell (top-right corner)

ax4 = plt.subplot(2,1,2) # fourth subplot in a 2-row by 1-column grid, occupying the second row entirely 

axes = [ax1, ax2, ax3, ax4] # groups all the subplot axes into a list named axes

# Making changes to all subplots
for ax in axes:
    ax.set_ylim(0.8, 1.7)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'],
                   alpha=0.3)
    ax.grid(alpha=0.5)    

# Creating first plot - Bush
ax1.plot(bush["Time"], bush["rolling_mean_30"], color = "blue")
ax1.set_xticks([10957, 11323, 11688, 12053, 12418, 12784, 13149, 13514, 13879, 14245])
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '', '2007', '', '2009'], alpha=0.3)
ax1.text(12418, 1.92, 'BUSH', fontsize=18, weight='bold', color = "blue")
ax1.text(12180, 1.8, '(2001-2009)', weight='bold', alpha=0.3)

# Creating second plot - Obama
ax2.plot(obama["Time"], obama["rolling_mean_30"], color = "red")
ax2.set_xticks([13879, 14245, 14610, 14975, 15340, 15706, 16071, 16436, 16801, 17167])
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '', '2015', '', '2017'], alpha=0.3)
ax2.text(15079, 1.92, 'OBAMA', fontsize=18, weight='bold', color='red')
ax2.text(14899, 1.8, '(2009-2017)', weight='bold', alpha=0.3)

# Creating third plot - Trump 
ax3.plot(trump["Time"], trump["rolling_mean_30"], color = "green")
ax3.set_xticks([17167, 17348, 17532, 17713, 17897, 18078, 18262, 18444, 18628])
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '', '2020', '', '2021'], alpha=0.3)
ax3.text(17567, 1.92, 'TRUMP', fontsize=18, weight='bold', color='green')
ax3.text(17485.0, 1.8, '(2017-2021)', weight='bold', alpha=0.3)

# Creating fourth plot - Bush, Obama, Trump 
ax4.plot(bush['Time'], bush['rolling_mean_30'], color='blue')
ax4.plot(obama['Time'], obama['rolling_mean_30'], color='red')
ax4.plot(trump['Time'], trump['rolling_mean_30'], color='green')
ax4.grid(alpha=0.5)
ax4.set_xticks([])

# Adding a title and subtitle
ax1.text(10757, 2.35, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(10757, 2.14, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''', fontsize=16)

plt.show()
```
